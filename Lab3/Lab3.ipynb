**Hint**

Puede ser √∫til los m√©todos ```.drop_duplicates()``` y ```.dropna()``` como tambi√©n utilizar queries con m√°scaras booleanas.


> Utilice la funci√≥n creada en el punto 1.1 para verificar sus resultados.

````python
#Funci√≥n que "limpia" el dataset
def clean_data(dataframe):
  #Se eliman datos duplicados
  df_cleaned = dataframe.drop_duplicates(subset = ["names"])

  #Se eliminan datos NaN
  df_cleaned = df_cleaned.dropna()

  #Se filtran notas fuera del rango [0,100]
  df_cleaned = df_cleaned[(df_cleaned.loc[:, "math score"] >= 0.0)
                              & (df_cleaned.loc[:, "math score"] <= 100.0)]
  df_cleaned = df_cleaned[(df_cleaned.loc[:, "reading score"] >= 0.0)
                              & (df_cleaned.loc[:, "reading score"] <= 100.0)]
  df_cleaned = df_cleaned[(df_cleaned.loc[:, "writing score"] >= 0.0)
                              & (df_cleaned.loc[:, "writing score"] <= 100.0)]

  return df_cleaned
````

### 3.1 Transformaci√≥n de notas a "Nota Chilena" [0.3 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intenci√≥n de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuaci√≥n:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuaci√≥n entregada, cree una funci√≥n que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [‚úì] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
#Funci√≥n que transforma la columna entregada del dataset a escala de notas chilena
def nota_chilena(dataframe_in, columna):
  dataframe_in[columna]=dataframe_in.loc[:,columna].apply(lambda x: round((x*0.06 + 1),2))
````

### 3.3 Reprobados y Destacados [0.9 Puntos, 0.3 cada DataFrame]

Contentos con su desempe√±o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a√±o y los que han tenido desempe√±o de destacado. Para esto se considerar√° que un alumno ha reprobado el a√±o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar√° destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci√≥n que permita obtener tres Dataframes:
- El primero deber√° poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser√° un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [‚úì] Obtener DataFrame con los alumnos que reprobaron el a√±o.
- [‚úì] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [‚úì] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):

    #DataFrame con los estudiantes reprobados (promedio menores que 4)
    df_rep = dataframe_in[(dataframe_in.loc[:,"GPA"]<4.0)]

    #DataFrame donde se pas√≥ a "R" los promedios menores a 4 en GPA del dataset original
    df_r = dataframe_in.copy()
    df_r.loc[(df_r["GPA"]<=4)] = "R"

    #DataFrame con los estudiantes destacados (notas mayores a 5.8 en todas las asignaturas)
    df_des = dataframe_in[(dataframe_in.loc[:,"math score"]>5.8)
                            & (dataframe_in.loc[:,"reading score"]>5.8)
                              & (dataframe_in.loc[:,"writing score"]>5.8)]
    df_des = df_des.sort_values('GPA', ascending = False)

    return df_rep, df_r, df_des
````

*Explicaci√≥n Caso Outer*

En este caso, se observa que se unen ambos dataframes completamente, causando un aumento en la cantidad de filas. Es probable que se deba a que el dataframe con las notas que se agregaron de historia y ciencias conten√≠a estudiantes que no estaban en el dataset original, causando que haya valores nulos en las columnas originales de estos estudiantes agregados. Por lo tanto, no corresponde al "merged" que se desea.

````python
#LEFT

df_left = pd.merge(
    left=df_grades_conc,
    right=df_other,
    on="names",
    sort=True,
    how="left",
)
````

````python
exploratory_data_analysis(df_left)
````

*Explicaci√≥n Caso Left*

En este caso, se observa que se unen el dataframe original completo y solo los estudiantes que ya exist√≠an en el dataset original, del dataset con notas extras. Se visualiza que no cambia la cantidad de filas y que tampoco existen valores nulos o duplicados, mostrando que solo se agregaron las notas faltantes. Este corresponde a uno de los "merged" adecuados a utilizar.

*OBS: Para este caso en espe√ßifico, se tiene que todos los estudiantes del dataframe original est√°n en el dataframe de notas extras. Si faltara uno o m√°s, estas filas tendr√≠an valores NaN en las notas de ciencias e historia, por lo que se tendr√≠a que utilizar otro "merge" para cumplir el objetivo (espec√≠ficamente Inner)*

````python
#RIGHT

df_right = pd.merge(
    left=df_grades_conc,
    right=df_other,
    on="names",
    sort=True,
    how="right",
)
````

````python
exploratory_data_analysis(df_right)
````

*Explicaci√≥n Caso Right*

En este caso, se observa un comportamiento similar al de outer. Al agregar todo el dataset con notas extras y solo la intersecci√≥n con el dataset original, permanecen las filas de los estudiantes en el dataset com notas extras que no ten√≠an informaci√≥n en los par√°metros del dataset original. Este no corresponde al "merge" que se desea utilizar.

````python
#INNER

df_inner = pd.merge(
    left=df_grades_conc,
    right=df_other,
    on="names",
    sort=True,
    how="inner",
)
````

*Explicaci√≥n Caso Inner*

Esto indica que corresponde al "merge" adecuado a utilizar.

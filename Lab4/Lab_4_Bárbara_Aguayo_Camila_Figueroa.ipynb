**Respuestas**

**¬øQu√© es Apache Spark y cu√°les son sus principales ventajas sobre Pandas?**

Apache Spark es un motor unificado para procesar datos a gran escala. Puede funcionar en centros f√≠sicos y en la nube.

Sus principales ventajas sobre Pandas son:
- Se desempe√±a de much√≠sima mejor forma al trabajar con vol√∫menes de datos grandes.
- Puede realizar operaciones distribuidas, siendo m√°s eficiente y r√°pido que Pandas, sobre todo utilizando gran cantidad de datos.
- Tiene una gesti√≥n de memoria mucho m√°s eficiente.

**¬øQu√© es un RDD en Spark? Describe una de sus principales caracter√≠sticas.**

Un Resilient Distributed Dataset (RDD) constituye el pilar fundamental de Apache Spark, siendo una estructura de datos distribuida e inmutable. Una de sus caracter√≠sticas principales es que los datos se encuentran divididos (o distribuidos) en particiones o nodos en el cl√∫ster.

**¬øQu√© tienen que ver con los dataframes?**

Los dataframes de PySpark corresponden a abstracciones de nivel superior de un RDD, es decir, son estructuras de datos m√°s simples de utilizar para los usuarios, dado que ofrecen una interfaz similar a las tablas de bases de datos con filas y columnas.

**Diferencia entre transformaciones y acciones en Spark. Proporciona ejemplos de cada una. ¬øQu√© ocurre internamente cuando se ejecuta una acci√≥n?**

Las transformaciones generan un nuevo dataframe o conjunto de datos distribuidos, pero sin modificar el dataframe original. Esto con el fin de realizar tareas como filtrar, mapear y agrupar los datos.

En cambio, las acciones son activadoras que desencadenan la ejecuci√≥n de las transformaciones necesarias para producir el resultado deseado.

**Explica la importancia del particionamiento en Spark y c√≥mo afecta el rendimiento del procesamiento de datos.**

El particionamiento en Spark permite un rendimiento eficiente en el procesamiento de datos, debido a que cada partici√≥n se procesa de forma independiente y distribuidas en el cl√∫ster. De esta forma, es posible realizar tareas y operaciones de forma paralela, lo que permite equilibrar la carga de trabajo entre los nodos del cluster, optimizando el rendimiento y los tiempos de ejecuci√≥n.

**¬øCu√°les son las funciones de Spark Driver y Spark Executor?**

Las distintas particiones son coordinadas a partir del Spark Driver, el cual tiene la funci√≥n de instanciar la SparkSession y comunicarse con el gestor del cluster para pedir recursos como memoria, CPU, entre otros.

Por otro lado, Spark Executor se encuentra en cada uno de los nodos del cluster y su funci√≥n es realizar las tareas asignadas por el conductor en el nodo respectivo y enviar los resultados al mismo.

**¬øQu√© es el Catalyst Optimizer en Apache Spark y cu√°l es su funci√≥n principal en la optimizaci√≥n de consultas SQL?**

El Catalyst Optimizer es una componente o herramienta fundamental de Spark SQL, que corresponde a una secci√≥n de Spark creada para ejecutar queries SQL. En esencia, corresponde a una biblioteca de √°rboles con instrucciones y c√≥mo manipularlo.

La funci√≥n principal de Catalyst Optimizer es mejorar el rendimiento de las consultas SQL al transformarlas en el plan m√°s eficiente para ejecutar las operaciones en los datos.

**Responda la siguiente pregunta: ¬øCu√°l/es columna/s tiene/n un tipo de dato que no es el adecuado y por qu√©?**

- Quantity debiera ser IntegerType -> porque no pueden haber cantidades de objetos comprados decimales

- Is An Alien debiera ser ByteType -> porque es 0 o 1 y no se necesitan decimales ni valores grandes

- Account Age Days debiera ser IntegerType -> La cantidad de d√≠as no pueden ser decimales

- Transaction Hour debiera ser IntegerType ->  debido a que la cantidad de horas no son decimales

````python
#Complete el c√≥digo entregado para cambiar el tipo de datos para la/s columna/s problem√°ticas.

#Se extrae funci√≥n de la Clase-Spark
from pyspark.sql.types import *


def cast_columns(df, cols_types):
    """
    Cambia el tipo de m√∫ltiples columnas en un DataFrame de Spark.

    Par√°metros:
    - df: DataFrame de Spark.
    - cols_types: Diccionario con nombres de columnas como claves y tipos de datos de Spark como valores.

    Retorna:
    - DataFrame de Spark con tipos de columnas modificados.
    """
    for col_name, new_type in cols_types.items():
        df = df.withColumn(col_name, df[col_name].cast(new_type))
    return df

# Ejemplo de uso
cols_to_cast = {"Quantity": IntegerType(), "Is An Alien": ByteType(), "Account Age Days": IntegerType(), "Customer Age": IntegerType(), "Transaction Hour": IntegerType()}
df = cast_columns(df, cols_to_cast)
df.printSchema()
````

````python
df.show(10)
````

````python
#Cuente la cantidad de datos nulos por variable.
#Recuerde que Spark no posee un m√©todo que le permita calcular directamente los nulos.

#Utilizamos m√©todo visto en clases
from pyspark.sql.functions import when, rand, col, sum

exprs = [sum(when(col(c).isNull(), 1).otherwise(0)).alias(c) for c in df.columns]

df.agg(*exprs).show()
````

````python
#Elimine datos nulos.

#Se eliminan cualquier fila que contenga al menos 1 dato nulo
df = df.na.drop(how="any")
````

````python
# Escriba su respuesta aqu√≠
#Agregar una columna llamada "Transaction bp" con el monto total de la transacci√≥n
#en bodoque pesos. Se considera que  x  d√≥lares equivalen a  log(48+|x**36|)  bodoque pesos.
from pyspark.sql.functions import abs, log

df=df.withColumn('Transaction bp', log(48+abs(df['Transaction Amount']**36)))
````

````python
df.show(5)
````

````python
#Crear una columna llamada "Transaction Month" con el mes en que se realiza una transacci√≥n.

from pyspark.sql.functions import month
df=df.withColumn('Transaction Month', month(df['Transaction Date']))
````

````python
df.show(5)
````

````python
#Crear la variable Type of purchase seg√∫n la cantidad de unidades vendidas de acuerdo a las siguientes categor√≠as.
#Compra minorista: 5 productos o menos.
#Compra mayorista: 6 produtos o m√°s.

df=df.withColumn('Type of purchase', when(df['Quantity']<=5, "Compra minorista").otherwise("Compra mayorista"))
````

````python
#Se observa que no hay ninguna compra mayorista
df.filter(df['Type of purchase']=='Compra mayorista').show()
df.filter(df['Quantity']>5).show()
````

````python
#Imprima los registros de compras hechas por alien√≠genas en el comecio mayorista. Utilice .filter().

df.filter(df['Is An Alien']==1).show()
````

````python
#Cuente la cantidad de compras realizadas por humanos y la cantidad de compras realizadas por alien√≠genas. Utilice .groupby().

df.groupby("Is An Alien").count().show()
````

````python
#Muestre una tabla con la recaudaci√≥n promedio por transacci√≥n para
#cada m√©todo de pago, tanto para humanos como alien√≠genas. Utilice pivot()

df.groupby("Payment Method").pivot("Is An Alien").mean("Transaction Amount").show()
````

Por ejemplo, UNBOUNDED PRECEDING se usa para indicar que el rango de la funci√≥n de ventana comienza desde la primera fila de la partici√≥n o del conjunto de resultados, lo cual es √∫til para calcular sumas acumulativas hasta la fila actual. - `VALUE PRECEDING` to `VALUE FOLLOWING`: Establece un rango espec√≠fico basado en valores antes y despu√©s de la fila actual.

STAT(COL1_NAME) OVER (PARTITION BY COL2_NAME ORDER BY COL3_NAME ROWS BETWEEN X PRECEDING AND CURRENT ROW)

Es un modelo que realiza Regresi√≥n Log√≠stica

¬øqu√© m√©todo del objeto predice la probabilidad?
M√©todo predict_proba

````python
# C√≥digo Aqu√≠

from pyspark.sql.functions import udf
from pyspark.sql.types import FloatType

#Crear una funci√≥n personalizada: Elabore una funci√≥n que prediga la probabilidad
#de fraude utilizando el √∫ltimo valor de la lista generada por el objeto serializado.
#Puede modificar el nombre de la funci√≥n para reflejar su prop√≥sito.

def predict_fraud(vals):
    vals = [vals]  # Convertir a array 2D
    predict_probs = loaded_object.predict_proba(vals)
    last = predict_probs[0][-1]
    #print(last)
    return float(last)

#Definir la funci√≥n UDF: Establezca la funci√≥n UDF con la funci√≥n personalizada que ha creado.
udf_function = udf(predict_fraud, FloatType())
````

````python
#Generar una nueva columna: A√±ada una nueva columna prediction a su DataFrame en
#Spark utilizando la funci√≥n UDF y muestre un ejemplo de c√≥mo se aplica.
from pyspark.sql.functions import array

columns_array = array('Transaction Amount', 'Quantity', 'Customer Age', 'Transaction Hour')


df_predictfraud = df.withColumn('Fraud Predictions', udf_function(columns_array))
````
